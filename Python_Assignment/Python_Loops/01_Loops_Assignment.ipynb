```python
# Write a program that takes a student's marks (0-100) as input and prints their grade according to the following criteria:

# 90-100: Grade A
# 80-89: Grade B
# 70-79: Grade C
# 60-69: Grade D
# Below 60: Grade F
# Also, validate that the marks are within the valid range (0-100).

Marks = int(input("Enter Marks from 0 to 100: "))
if Marks < 0 or Marks > 100:
    print("Invalid Marks")
elif Marks >= 90:
    print(f"Student Marks: {Marks}, Grade A")
elif Marks >= 80:
    print(f"Student Marks: {Marks}, Grade B")
elif Marks >= 70:
    print(f"Student Marks: {Marks}, Grade C")
elif Marks >= 60:
    print(f"Student Marks: {Marks}, Grade D")
else:
    print(f"Student Marks: {Marks}, Grade F")
```

```python
# Write a program to check if a given year is a leap year or not.

Year = int(input("Enter Year:  "))
if (Year %4==0 and Year %100 !=0) or (Year %400==0):
    print(f"{Year} is a Leap Year")
else:
    print(f"{Year} is not a Leap Year")
```

```python

# Write a program to determine the type of triangle (equilateral, isosceles, or scalene) based on the lengths of its sides provided by the user.

Side1 = int(input("side 1 of triangle: "))
Side2 = int(input("side 2 of triangle: "))
Side3 = int(input("side 3 of triangle: "))

if Side1==Side2==Side3:
    print(f"Side {Side1} is an Equilateral Triangle")
elif Side1==Side2 or Side2==Side3 or Side1==Side3:
    print(f"Side {Side1}, {Side2}, and {Side3} is an Isosceles Triangle")
else:
    print(f"Side {Side1}, {Side2}, and {Side3} is a Scalene Triangle")
```

```python
# Write a program that calculates BMI (Body Mass Index) and categorizes it
# Formula: BMI = weight (kg) / (height (m))Â²

# Categories:
# Below 18.5: Underweight
# 18.5-24.9: Normal weight
# 25-29.9: Overweight
# 30 and above: Obese

weight = float(input("enter the weight:"))
height = float(input("enter the height:"))

bmi = weight/(height * height)

if bmi < 18.5:
    print("under weight")
elif bmi >= 18.5 and bmi <= 24.9:
    print("normal weight")
elif bmi >= 25 and bmi <=29.9:
    print("overweight")
else:
    print("obese")
```
