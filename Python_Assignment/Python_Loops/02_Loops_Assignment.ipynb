```python
# 1.Count how many vowels are in a given string.

String = input("Enter a String:  ")
vowels = "aeiouAEIOU"
count = 0
for i in String:
    if i in vowels:
        count+=1
print(f"Number of Vowels in the String is: {count}")
```

```python

# 2.Reverse a number using ONLY a for loop (no slicing, no string reverse).

num = int(input("Enter numbers:  "))
rev = 0
while num>0:
    value = num%10
    rev = rev*10 + value
    num //= 10
print(f"Reverse of the number is: {rev}")
```

```python
# 3.Find the sum of digits of a number using a loop.

num = int(input("Enter number: "))
total = 0
for i in range(len(str(num))):
    last_digit = num % 10
    total += last_digit
    num = num // 10
print(f"total sum of digits is: {total}")
```

```python
# 5. Factorial of a number

num = int(input("Enter number: "))
fact = 1
while num>0:
    fact *= num
    num -= 1
print(f"Factorial of the number is: {fact}")
```

```python
# 6. Check whether a number is prime using a for loop

num = int(input("Enter number: "))
def prime(num):
    if num > 1:
        for i in range(2,num):
            if (num % i) == 0:
                print(num, "is not a prime number")
                break
        else:
            print(num, "is a prime number")
    else:
        print(num, "is not a prime number")
if prime(num):
    print(num, "is a prime number")
```

```python
#Question 9: Find the largest number in a list Example: [4, 7, 1, 9, 3]

list1 = [4, 7, 1, 9, 3]
greater = list1[0]
for i in list1:
    if i > greater:
        greater = i
        
print(f"Greater number is {greater}" )
```
