### Lambda Function

A lambda function is a small, anonymous (unnamed) function defined using the lambda keyword.

### syntax

#### lambda arguments: expression

- Its can take any numbers of arguments
- It can have only one expression 
- It automaticallly rturn the result of the expresssion 
- No retrun keyword required 

```python
# Normal Function

def add(a,b):
    return a+b
```

```python
addition = add(3,4)
```

```python
addition
```

```python
# Lambda function

addition1 = lambda a,b : a+b

addition1(3,4)
```

```python
### lambda with single arguement

# square a number 

square = lambda a : a**2
square(2)
```

```python
square(7)
```

```python
## lambda with multiple arguments 


Area = lambda length , width : length*width
Area(10,20)
```

```python
Area(5,4)
```

```python
# lambda with conditional 

is_even = lambda x : "Even" if x%2==0 else "Odd"
is_even(20)
```

```python
is_even(13)
```

```python
# Changing lower case to upper case using normal functions

def to_upper(s):
    return s.upper()
```

```python
to_upper("adefjalnfl")
```

```python
# Same using lambda function

to_upper = lambda s : s.upper()
```

```python
to_upper("eijfba")
```

```python
to_lower = lambda s : s.lower()
```

```python
to_lower("EOIFHW")
```

```python
def string_reverse(s):
    return s[::-1]
```

```python
string_reverse("reverse this")
```

```python
# Reverse String using Lambda function

string_reverse = lambda s : s[::-1]
```

```python
string_reverse("Hello")
```

### filter()

filter(function, iterable)

It filters elements from an iterable (like list, tuple).

Keeps only those elements for which the function returns True.

Returns a filter object (convert to list if needed).

```python
# Filter Even numbers

numbers = [1, 2, 3, 4, 5, 6]

even_number = list(filter(lambda x : x%2==0 , numbers ))
```

```python
even_number
```

```python
# Filter Odd Numbers

numbers = [10, 15, 26, 31, 33, 50, 61]

Odd_numbers = list(filter(lambda x : x%2!=0 , numbers))
```

```python
Odd_numbers
```

```python
# Filter Positive numbes

numbers = [-1, 3, 5, -3, 6, -6, 7]

Positive_numbers = list(filter(lambda x : x>0 , numbers))
```

```python
Positive_numbers
```

```python
# Filter Strings with Length Greater than 5

words = ["apple", "banana", "kiwi", "watermelon"]

result = list(filter(lambda w : len(w)>5 , words))
```

```python
result
```

```python
# Filter Names Starting with 'A'

names = ["Anil", "Ajay", "Ravi", "Arjun"]

Startswith_A = list(filter(lambda s : s.startswith("A") , names))
```

```python
Startswith_A
```

```python
#  Remove Empty Strings from List

data = ["hello", "" , "world", "" , "python"]

Remove_empty = list(filter(lambda s : s !="" , data))
```

```python
Remove_empty
```

```python
#  Filter Numbers Greater than 50

numbers = [10, 55, 60, 23, 89, 45]

result = list(filter(lambda x : x>50 , numbers))
```

```python
result
```
