### what is map()? 

- applies a fucniton to every element 
- returns a map object (convert to list if needed)

```python
## passing sequence datatypes wiht lambda 

# map()

numbers = (2,4,5,6,7,9)

square = list(map(lambda s : s**2,numbers))
```

```python
square
```

```python
celcius = (0, 10, 20, 30, 40)
fahrenhiet = list(map(lambda c : (c*9/5) + 32 , celcius))
```

```python
fahrenhiet
```

```python
## syntax for map 

# list(map(lambda x: x, sequence))
```

```python
## map(function, iterable)
```

```python
names = ["lohit", "param", "rahul"]

to_upper = list(map(lambda s : s.upper() , names))
```

```python
to_upper
```

```python
numbers = [3,4,5,7,9,8]
cube = list(map(lambda n : n**3 , numbers))
```
